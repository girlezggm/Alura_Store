### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



# **Ingreso total por cada tienda**

```python
# Ingresos totales por cada tienda

# Ingresos totales por tienda
ingresos_tienda1 = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()

print(f"Ingresos totales tienda 1: {ingresos_tienda1}")
print(f"Ingresos totales tienda 2: {ingresos_tienda2}")
print(f"Ingresos totales tienda 3: {ingresos_tienda3}")
print(f"Ingresos totales tienda 4: {ingresos_tienda4}")
```

```python
# Datos de ingresos por tienda (reemplaza con tus valores reales)
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura
plt.bar(tiendas, ingresos, color=['skyblue', 'lightcoral', 'lightgreen', 'lightblue'])
plt.xlabel('Tiendas')
plt.ylabel('Ingresos totales')
plt.title('Ingresos totales por tienda')
plt.grid(axis='y', alpha=0.5) # Agregar una cuadrícula suave al eje y
plt.show()
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y contar ventas
ventas_categoria_tienda = tienda['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda.columns = ['Categoría del Producto', 'Cantidad Vendida']

ventas_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda2.columns = ['Categoría del Producto', 'Cantidad Vendida']

ventas_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda3.columns = ['Categoría del Producto', 'Cantidad Vendida']

ventas_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda4.columns = ['Categoría del Producto', 'Cantidad Vendida']
```

```python
# Mostrar resultados
print("Tienda  - Ventas por Categoría:")
print(ventas_categoria_tienda)
print("\nTienda 2 - Ventas por Categoría:")
print(ventas_categoria_tienda2)
print("\nTienda 3 - Ventas por Categoría:")
print(ventas_categoria_tienda3)
print("\nTienda 4 - Ventas por Categoría:")
print(ventas_categoria_tienda4)
```

```python
# grafico de pastel por cada tienda

def plot_pie_chart(df, title):
    plt.figure(figsize=(8, 6))
    plt.pie(df['Cantidad Vendida'], labels=df['Categoría del Producto'], autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

plot_pie_chart(ventas_categoria_tienda, 'Ventas por Categoría - Tienda 1')
plot_pie_chart(ventas_categoria_tienda2, 'Ventas por Categoría - Tienda 2')
plot_pie_chart(ventas_categoria_tienda3, 'Ventas por Categoría - Tienda 3')
plot_pie_chart(ventas_categoria_tienda4, 'Ventas por Categoría - Tienda 4')
```

# 4. Productos más y menos vendidos

```python
# productos fueron los más vendidos y los menos vendidos en cada tienda

def productos_mas_menos_vendidos(tienda, nombre_tienda):
    ventas_producto = tienda['Producto'].value_counts().reset_index()
    ventas_producto.columns = ['Producto', 'Cantidad Vendida']

    mas_vendido = ventas_producto.iloc[0]
    menos_vendido = ventas_producto.iloc[-1]

    print(f"Tienda {nombre_tienda}:")
    print(f"Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad Vendida']} ventas)")
    print(f"Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad Vendida']} ventas)\n")

    # Gráfico de barras para productos más y menos vendidos
    plt.figure(figsize=(10, 6))
    plt.bar(ventas_producto['Producto'], ventas_producto['Cantidad Vendida'], color='lightcoral')
    plt.title(f'Cantidad de Productos Vendidos - {nombre_tienda}')
    plt.xlabel('Nombre del Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=90)  # Rotar etiquetas del eje x
    plt.tight_layout() # Ajusta el diseño
    plt.show()


productos_mas_menos_vendidos(tienda, "1")
productos_mas_menos_vendidos(tienda2, "2")
productos_mas_menos_vendidos(tienda3, "3")
productos_mas_menos_vendidos(tienda4, "4")
```

# 5. Envío promedio por tienda

```python
# Calcular el coste medio de envío de cada tienda
envio_promedio_tienda = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Mostrar los resultados
print(f"Envío promedio en Tienda: {envio_promedio_tienda:.2f}")
print(f"Envío promedio en Tienda 2: {envio_promedio_tienda2:.2f}")
print(f"Envío promedio en Tienda 3: {envio_promedio_tienda3:.2f}")
print(f"Envío promedio en Tienda 4: {envio_promedio_tienda4:.2f}")

# Crear un gráfico de barras para visualizar los gastos de envío medios
tiendas = ['Tienda', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios_envio = [envio_promedio_tienda, envio_promedio_tienda2, envio_promedio_tienda3, envio_promedio_tienda4]

plt.figure(figsize=(8, 6))
plt.bar(tiendas, promedios_envio, color='lightgreen')
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

# Agrupar por región y calcular métricas de rendimiento
performance_by_region = all_stores.groupby('Lugar de Compra')['Precio'].agg(['sum', 'mean', 'count'])

# Identificar regiones con rendimiento superior o inferior
threshold_performance = performance_by_region['mean'].mean()
high_performing_regions = performance_by_region[performance_by_region['mean'] > threshold_performance]
low_performing_regions = performance_by_region[performance_by_region['mean'] < threshold_performance]

print("Regiones con alto rendimiento:")
print(high_performing_regions)

print("\nRegiones con bajo rendimiento:")
print(low_performing_regions)


# Análisis de patrones geográficos
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
all_stores['Region_encoded'] = le.fit_transform(all_stores['Lugar de Compra'])


plt.figure(figsize=(10, 8))
heatmap_data = all_stores.groupby(['lat', 'lon'])['Region_encoded'].mean().unstack()
sns.heatmap(heatmap_data, cmap='YlOrRd')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Mapa de Calor de Regiones por Ubicación')
plt.show()

```
